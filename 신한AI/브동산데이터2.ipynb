```python
import pandas as pd
```

```python
seoul_df = pd.read_csv('/content/drive/MyDrive/test/서울_인코딩_임시.csv')
gg_df = pd.read_csv('/content/drive/MyDrive/test/경기도_인코딩_임시.csv')
```

```python
s_coll=seoul_df.columns[-2:].to_list()
s_coll2=seoul_df.columns[:-2].to_list()
seoul_df=seoul_df[s_coll+s_coll2]
seoul_df = seoul_df.drop(['시/군/구','동'],axis=1)
```

```python
g_coll=gg_df.columns[-2:].to_list()
g_coll2=gg_df.columns[:-2].to_list()
gg_df = gg_df[s_coll+s_coll2]
gg_df = gg_df.drop(['시/군/구','동'],axis=1)
```

```python
seoul_df['종가'] = seoul_df['종가'].astype('str')
seoul_df['종가'] = seoul_df['종가'].str.replace(',','').astype('float')
gg_df['종가'] = gg_df['종가'].astype('str')
gg_df['종가'] = gg_df['종가'].str.replace(',','').astype('float')
```
